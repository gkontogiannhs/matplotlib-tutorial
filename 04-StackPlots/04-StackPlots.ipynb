```python
# import out beloved library
from matplotlib import pyplot as plt
# make sketch editable for the notebook
%matplotlib notebook
```

```python
# choose custom style
plt.style.use('fivethirtyeight')
```

### Creating random data suitable for stack plots

```python
date = [2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]
```

```python
from random import randint
```

```python
x_len = len(date)

solar = [randint(1000, 3000) for _ in range(x_len)]
wind = [randint(4000, 10000) for _ in range(x_len)]
natural_gas = [randint(10000, 50000) for _ in range(x_len)]
large_hydro = [randint(10000, 20000) for _ in range(x_len)]
```

### Creating a stack plot

```python
# labels to be used for the legend
labels = ['natural_gas', 'solar', 'wind', 'large_hydro']

# Similar to plot and bar function, stackplot works the same way. First we pass the x axis values which 
# in our situation is date variable and then as many data arguments as we wish. 
plt.stackplot(date, natural_gas, solar, wind, large_hydro, labels=labels)

# y data variables can also be passed as a multidimensional variable like the statement below
# plt.stackplot(date, [natural_gas, solar, wind, large_hydro], labels=labels)

# title, axis labeling, legend and decoration
plt.title('Electicity Supply Per Energy Source', fontsize=10)
plt.xlabel('date', fontsize=12)
plt.ylabel('GWh', fontsize=12)
plt.tight_layout()
plt.legend(loc=(0.05, .8), prop={'size':7})
plt.show()
```
