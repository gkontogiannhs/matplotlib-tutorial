## Pie charts 

```python
from matplotlib import pyplot as plt
%matplotlib notebook
```

```python
plt.style.use('fivethirtyeight')
```

```python
# data to use
categories = [10, 20, 30, 40]

colors = ['#008fd5', '#fc4f30', '#e5ae47', '#6d904f']
labels = ['Ten', 'Twenty', 'Thirty', 'Forty']

# pie function creates the pie
plt.pie(categories, labels=labels, colors=colors)
plt.title('Pie Chart')
plt.tight_layout()
plt.show()
```

## Decorations
### wedgeproperties - https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.patches.Wedge.html

```python
# pie function creates the pie
plt.pie(categories, labels=labels, colors=colors, 
       wedgeprops={'edgecolor':'#000000'})
plt.title('Pie Chart')
plt.tight_layout()
plt.show()
```

### Exploding and shadow effect

```python
explode = [0, 0, 0.1, 0]
# pie function creates the pie
plt.pie(categories, labels=labels, colors=colors, shadow=True,
       wedgeprops={'edgecolor':'#000000'}, explode=explode)
plt.title('Pie Chart')
plt.tight_layout()
plt.show()
```

### Rotation - startangle attribute

```python
plt.pie(categories, labels=labels, colors=colors, shadow=True,
       wedgeprops={'edgecolor':'#000000'}, explode=explode, startangle=55)
plt.title('Pie Chart')
plt.tight_layout()
plt.show()
```

### Percentages per chunk

```python
plt.pie(categories, labels=labels, colors=colors, shadow=True,
        wedgeprops={'edgecolor':'#000000'}, explode=explode,
        startangle=55, autopct='%1.1f%%')
plt.title('Pie Chart')
plt.tight_layout()
plt.show()
```
