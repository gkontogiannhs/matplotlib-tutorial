# Creating and customizing our first plots

### Official Documentation

### https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

```python
# importing matplotlib
from matplotlib import pyplot as plt
%matplotlib notebook
```

## Default line plot with single arguement

```python
real_numbers = [0.5, 1.0, 1.5, 2.0, 2.5]
real_numbers
```

```python
plt.plot(real_numbers)
plt.show()
```

```python
# So, from the above graph we conclude that plot function treats single argument as the y function while
# it generates integers for the x axis until it matches the length of y.
```

## Simple Line plot

```python
# x_axis = [x for x in range(0, 10)]
x_axis = list(range(0, 10))
x_axis
```

```python
# y = ax + b, a = b = 1 
# y_axis = [y for y in range(0, 10)]
y_axis = list(range(0, 10))
y_axis
```

```python
# we can plot our data using plot function by passing x and y axis
plt.plot(x_axis, y_axis)
plt.show()
```

## Line plot with axis labels and title

```python
# plot our data
plt.plot(x_axis, y_axis)

# put custom axis labels
plt.xlabel('x')
plt.ylabel('y')

# put title
plt.title('Linear Function')
plt.show()
```

## Multiple graphs in the same plot with legend

```python
y1 = y_axis[:]
y2 = [n*n for n in x_axis]
```

```python
# use plot function as many time as u want before show funciton
plt.plot(x_axis, y1, label='linear')
plt.plot(x_axis, y2, label='quadratic')
# other way for legend
# plt.legend(['linear', 'quadratic'])

# labels
plt.xlabel('x')
plt.ylabel('y')

# title
plt.title('Linear and quadratic function')
plt.legend() # shows the legend
plt.show()
```

##  Format String

```python
# A format string consists of a part for color, marker and line:
# fmt = '[color][marker][line]'    
```

```python
# plt.plot(x_axis, y1, 'm--', label='linear')
# ploting
plt.plot(x_axis, y1, label='linear', color='m', linestyle=':', marker='s')
plt.plot(x_axis, y2, label='quadratic', color='#0000FA', linestyle='--', marker='o')

# labels
plt.xlabel('x')
plt.ylabel('y')

# title
plt.title('Linear and quadratic function')
plt.legend()
plt.show()
```

## Linewidth, layout and grid

```python
y3 = [n*n*n for n in range(0, 10)]
```

```python
# ploting
plt.plot(x_axis, y1, label='linear', color='m', linestyle=':', marker='s', linewidth=2)
plt.plot(x_axis, y2, label='quadratic', color='#0000FA', linestyle='--', marker='o', linewidth=2)
plt.plot(x_axis, y3, label='cubic', color='#ADB345', linestyle='-', marker='.')

# labels
plt.xlabel('x')
plt.ylabel('y')

# title
plt.title('Linear and quadratic function')
plt.legend()
plt.tight_layout() # helps with padding
plt.grid(True) # enables the grid plot
plt.show()
```
